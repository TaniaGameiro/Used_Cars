```python
import numpy as np
import os
import pandas as pd
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy

get_ipython().run_line_magic('matplotlib', 'inline')
import seaborn as sns
sns.set(color_codes=True)
```

```python
UsedCars = pd.read_csv("/Users/Nora/Desktop/true_car_listings.csv")
```

Going to do a quick check to make sure we uploaded the right datset with the adjusted columns. 

```python
UsedCars.head()
```

Now lets move on to see which Make we have the most listings for available 

```python
UsedCars.Make.value_counts()
```

The most available vehicle is clearly the FORD by a large percentage. Now within the FORD category, which models are the most available. 

```python
UsedCars[UsedCars.Make == "Ford"].Model.value_counts()
```

The most available vehicle is the FORD F-150AWD. 

Let's sort the data by year to see what the most recent vehicles included in this list are. 

```python
UsedCars.sort_values(by=["Year"], ascending=False)
```

Let's look up any car that cost less than 10,000$ while sorting by year. 

```python
UsedCars[UsedCars['Price' ] <= 10000]
```

We can group the data to see the mean and rounded values. 

```python
UsedCars.groupby(by='Year')['Mileage'].mean().round()
```

```python
UsedCars.groupby(by='Year')['Price'].mean().round()
```

```python
UsedCars.groupby(by='Make')['Price'].mean().round()
```

## Pivot Table

Let's do a pivot table to see the data clearer

```python
UsedCars.pivot_table(index='Price', columns='Year', values='Mileage', aggfunc=np.mean)
```

Let's plot out the quantity of vehicles vs the year of the car. This way we can see which car year is most available. 

## Data Visualization

```python
UsedCars['Year'].value_counts().head(15).plot.bar()
plt.xlabel('Car Year')
plt.ylabel('Qty of Vehicles')
print("Cars Available based on Car Year")
```

```python
# Scatter plot
plt.plot(UsedCars['Price'], UsedCars['Year'], 'o')
plt.xlabel('Price')
plt.ylabel('Year')
print("Car Price Vs Car Year")
```

```python
# Scatter plot
plt.plot(UsedCars['Price'], UsedCars['Mileage'], 'o')
plt.xlabel('Price')
plt.ylabel('Mileage')
print("Car Price Vs Car Mileage")
```

```python
# Scatter plot multiple series
plt.figure(figsize=(12,6))
plt.plot(UsedCars['Price'], UsedCars['Year'], 'o', color='purple', alpha=0.2)
plt.xlabel('Price')
plt.ylabel('Year')
print("Car Price Vs Car Year")
```

```python
# Scatter plot multiple series
plt.plot(UsedCars['Price'], UsedCars['Mileage'], 'o', color='red', alpha=0.2)
plt.xlabel('Price')
plt.ylabel('Mileage')
print("Car Price Vs Car Mileage")
```

```python
# Histogram
plt.hist(UsedCars['Year'], bins=10)
plt.show()
```
